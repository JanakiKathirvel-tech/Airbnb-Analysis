```python
import pandas as pd
dataframe1 = pd.read_excel('output.xlsx')
```

```python
noofreviews_roomtype= pd.DataFrame(dataframe1, columns=("room_type", "number_of_reviews"))
```

```python
import plotly.express as px
import matplotlib.pyplot as plt
```

```python
lt= noofreviews_roomtype[["room_type", "number_of_reviews"]]
lt1= lt.groupby("room_type")["number_of_reviews"].sum().sort_values(ascending= False)
lt2= pd.DataFrame(lt1).reset_index().head(10)
fig_pie_1= px.pie(data_frame=lt2, names= "room_type", values="number_of_reviews", hover_data= "number_of_reviews",
                width=1000,title=f"RoomTypewise Number of reviews",hole=0.5, color_discrete_sequence= px.colors.sequential.Magenta_r)
```

```python
fig= px.bar(noofreviews_roomtype, x= "room_type", y= "number_of_reviews", 
                        color=px.colors.sequential.Greens_r, width= 600, 
                        title= f"Number of Reviews RoomType wise",height= 500)
```
